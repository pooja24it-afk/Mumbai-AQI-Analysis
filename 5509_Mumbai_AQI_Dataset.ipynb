```python
plt.figure(figsize=(10,5))
plt.hist(df["AQI"], bins=15, rwidth=0.95)
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.dtypes
```

```python
plt.figure(figsize=(8,5))
df["AQI"].plot(kind="density", color="red")
plt.xlabel("AQI")
plt.ylabel("Density")
plt.title("AQI Density Distribution of Mumbai")
plt.show()
```

```python
monthly_aqi = df.groupby("Month")["AQI"].mean()

plt.figure(figsize=(8,5))
plt.plot(monthly_aqi.index, monthly_aqi.values, marker='o')
plt.xlabel("Month")
plt.ylabel("Average AQI")
plt.title("Monthly Average AQI of Mumbai")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(df["AQI"])
plt.ylabel("AQI")
plt.show()
```

```python
df.groupby("Month")["AQI"].mean().plot(kind="bar", figsize=(8,5))
plt.xlabel("Month")
plt.ylabel("Average AQI")
plt.show()

```
